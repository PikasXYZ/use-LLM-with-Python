```python
#@title import
import anthropic
import base64
from google.colab import userdata
from IPython.display import Image, display
```

```python
#@title setting
client = anthropic.Anthropic(api_key=userdata.get('ANTHROPIC_API_KEY')) # get api from https://console.anthropic.com/settings/keys
```

```python
#@title chat w/ history
message = client.messages.create(
    model="claude-3-opus-20240229",
    max_tokens=1000,
    temperature=0,
    system="You're Pikas' powerful assistant",
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "Give me some positive words starting w/ P"},
                ]
        }
    ]
)
print(message.content[0].text)
```

```python
#@title chat w/ history & images
def encode_image(image_path): # encoding function
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')

message = client.messages.create(
    model="claude-3-opus-20240229",
    max_tokens=1000,
    temperature=0.1,
    system="You're Pikas' powerful assistant",
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "image",
                    "source": {
                        "type": "base64",
                        "media_type": "image/jpeg",
                        "data": encode_image("pikas_head.jpg")
                    }
                },
                {
                    "type": "text",
                    "text": "What's in this image?"
                },
                {
                    "type": "image",
                    "source": {
                        "type": "base64",
                        "media_type": "image/jpeg",
                        "data": encode_image("pikas_head2.jpg")
                    }
                },
                {
                    "type": "text",
                    "text": "What's the difference between this new image and the previous one?"
                }
            ]
        }
    ]
)
print(message.content[0].text)
display(Image("pikas_head.jpg"))
display(Image("pikas_head2.jpg"))
```
