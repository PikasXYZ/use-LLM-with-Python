```python
#@title import
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch
```

```python
#@title settings
MODEL_NAME = 'Qwen/Qwen2-1.5B-Instruct' # find other MODEL_NAME in https://huggingface.co/models?pipeline_tag=text-generation
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME) # download tokenizer
model = AutoModelForCausalLM.from_pretrained(MODEL_NAME) # download model
```

```python
#@title Generating text
def generate(prompt, temperature=0.7, max_new_tokens=50, do_sample=True, top_k=50, top_p=0.95, no_repeat_ngram_size=2, num_return_sequences=1) -> str:
    global tokenizer, model
    input = tokenizer(prompt, return_tensors='pt')
    output = model.generate(
        **input,
        do_sample=do_sample,
        top_k=top_k,
        top_p=top_p,
        temperature=temperature,
        num_return_sequences=num_return_sequences,
        no_repeat_ngram_size=no_repeat_ngram_size,
        max_new_tokens=max_new_tokens
    )
    return tokenizer.decode(output[0], skip_special_tokens=True)
print(generate('Hello, my name is Pikas.', do_sample=True))
```

```python
#@title And some chat templates of tokenizer add special tokens among contents
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2-7B-Instruct")
messages = [
  {"role": "user", "content": "Hello, how are you?"},
  {"role": "assistant", "content": "I'm doing great. How can I help you today, Pikas?"},
  {"role": "user", "content": "I'd like to show off how chat templating works!"},
]
tokenizer.apply_chat_template(messages, tokenize=False)
```

```python
#@title use transformers.pipeline for chat cuz it calls apply_chat_template when the model has a chat template
model_name = "Qwen/Qwen2-7B-Instruct"
pipe = pipeline("text-generation", model_name)

def chat(messages:list, max_new_tokens=128) -> str:
  global pipe
  return pipe(messages, max_new_tokens=max_new_tokens)[0]['generated_text'][-1]['content']

messages = [
    {"role": "system", "content": "You are Pikas' dog that speaks english ends with \"woof woof\""},
    {"role": "user", "content": "Wanna eat dinner?"},
 ]
print(chat(messages))
```

#Some LLM can use functions as tools!

```python
#@title Set a model able to use tools
model_name = "NousResearch/Hermes-2-Pro-Llama-3-8B"
tokenizer = AutoTokenizer.from_pretrained(model_name, revision="pr/13")
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.bfloat16)
```

```python
#@title Make some tools
import random
def get_height(name:str) -> str:
  """ Get one's height in cm """
  return str(180) if name == "Pikas" else str(random.randint(150, 179))

def get_weight(name:str) -> str:
  pass

def get_gender(name:str) -> str:
  pass

def get_age(name:str) -> str:
  pass

def get_IQ(name:str) -> str:
  """ Get one's IQ """
  return str(200) if name == "Pikas" else str(random.randint(1, 199))

tools = [
  {
    "type": "function",
    "function": {
      "name": "get_height",
      "description": "Get one's height in cm",
      "parameters": {
        "type": "object",
        "properties": {
          "name": {
            "type": "string",
            "description": "The name of the person",
          },
        },
        "required": ["name"],
      },
    },
  },
  {
    "type": "function",
    "function": {
      "name": "get_IQ",
      "description": "Get one's IQ",
      "parameters": {
        "type": "object",
        "properties": {
          "name": {
            "type": "string",
            "description": "The name of the person",
          },
        },
        "required": ["name"],
      },
    },
  },
]
```

```python
#@title settings
messages = [
  {"role": "system", "content": "You are a powerufl bot that knows everyone's height, weight, gender, age and IQ."},
  {"role": "user", "content": "Hey, what's the height & IQ of Amy?"}
]
```

```python
#@title Let's see how LLM calls funtcions
inputs = tokenizer.apply_chat_template(messages, chat_template="tool_use", tools=tools, add_generation_prompt=True, return_dict=True, return_tensors="pt")
inputs = {k: v.to(model.device) for k, v in inputs.items()}
out = model.generate(**inputs, max_new_tokens=128)
print(tokenizer.decode(out[0][len(inputs["input_ids"][0]):-1]))
```

```python
#@title Append tool_calls to messages...
messages.append({"role": "assistant", "tool_calls": [{"id": "pikasfunction1", # Random ID, should be unique for each tool call
                            "type": "function",
                            "function": {"arguments": {"name": "Amy"}, "name": "get_height"}}]})
messages.append({"role": "assistant", "tool_calls": [{"id": "pikasfunction2", # Random ID, should be unique for each tool call
                            "type": "function",
                            "function": {"arguments": {"name": "Amy"}, "name": "get_IQ"}}]})
```

```python
#@title And see output!
inputs = tokenizer.apply_chat_template(messages, chat_template="tool_use", tools=tools, add_generation_prompt=True, return_dict=True, return_tensors="pt")
inputs = {k: v.to(model.device) for k, v in inputs.items()}
out = model.generate(**inputs, max_new_tokens=128)
print(tokenizer.decode(out[0][len(inputs["input_ids"][0]):-1]))
```

```python
#@title Turn whole process into a function
def chat_with_tools(messages=[{"role": "system", "content": "You're a powerufl assistant."}], tools=[]):
    global tokenizer, model

    final_response = "default response"
    try:
      if tools:
        inputs = tokenizer.apply_chat_template(messages, chat_template="tool_use", tools=tools, add_generation_prompt=True, return_dict=True, return_tensors="pt")
        inputs = {k: v.to(model.device) for k, v in inputs.items()}
        out = model.generate(**inputs, max_new_tokens=128)
        response_text = tokenizer.decode(out[0][len(inputs["input_ids"][0]):-1])

        # get tool_calls
        tool_calls = []
        for line in response_text.split('</tool_call>'):
            if '<tool_call>' in line:
                tool_call_json = line.replace('<tool_call>', '').strip()
                tool_calls.append(eval(tool_call_json))

        # update messages w/ processed tool_calls
        for call in tool_calls:
            func_name = call['name']
            arguments = call['arguments']
            tool_result = globals()[func_name](**arguments)
            messages.append({
                    "role": "assistant",
                    "tool_calls": [{
                            "id": f"pikasfunction_{func_name}",
                            "type": "function",
                            "function": call,
                            "result": tool_result
                            }]
                    })

      # get final response w/ updated messages
      inputs = tokenizer.apply_chat_template(messages, chat_template="tool_use" if tools else None, tools=tools, add_generation_prompt=True, return_dict=True, return_tensors="pt")
      inputs = {k: v.to(model.device) for k, v in inputs.items()}
      out = model.generate(**inputs, max_new_tokens=128)
      final_response = tokenizer.decode(out[0][len(inputs["input_ids"][0]):-1])

    except Exception as e:
      print(str(e))
      print(f"{response_text = }\n{tool_calls = }\n{messages = }")

    return final_response
```

```python
#@title test
messages = [
    {"role": "system", "content": "You are a powerful bot that knows everyone's height, weight, gender, age and IQ."},
    {"role": "user", "content": "Hey, what's the height & IQ of Amy?"}
]

response = chat_with_tools(messages, tools)
response
```
